```python
"""
Реализовать классы Espresso, Americano, Cappuccino, задав им базовые названия, описания и цены.
Дополнить декораторы (MilkDecorator, SyrupDecorator, SugarDecorator), чтобы они корректно 
изменяли описание и стоимость напитка.
Добавить стратегию FoamyBrew, которая добавляет пенку к напитку (+0.75$ к стоимости).
Реализовать сборку сложного заказа (например, 2 декорированных напитка с разными стратегиями) через OrderBuilder.
Убедиться, что CoffeeMachine является синглтоном: создать несколько экземпляров и проверить, что они ссылаются на один объект.
Продемонстрировать работу механизма Observer: создайте 2–3 клиента, подпишите их на уведомления, оформите несколько заказов
и убедитесь, что клиенты получают сообщения.
В комментариях к коду кратко объясните, почему в каждом месте выбран именно этот паттерн 
(например: «Используем декоратор, чтобы гибко добавлять ингредиенты без изменения базовых классов напитков»).
"""

from abc import ABC, abstractmethod
from typing import List, Optional

# --- Интерфейс кофейного напитка ---

class ICoffee(ABC):
    @abstractmethod
    def get_description(self) -> str:
        pass

    @abstractmethod
    def get_cost(self) -> float:
        pass


# --- Базовые напитки (реализовать) ---

class Espresso(ICoffee):
    def __init__(self):
        # TODO: задать базовое описание и стоимость
        pass

    def get_description(self) -> str:
        # TODO
        return ""

    def get_cost(self) -> float:
        # TODO
        return 0.0


class Americano(ICoffee):
    # Аналогично Espresso
    pass


class Cappuccino(ICoffee):
    # Аналогично Espresso
    pass


# --- Абстрактная фабрика ---

class CoffeeFactory:
    def create_espresso(self) -> ICoffee:
        return Espresso()

    def create_americano(self) -> ICoffee:
        return Americano()

    def create_cappuccino(self) -> ICoffee:
        return Cappuccino()


# --- Декораторы (реализовать) ---

class MilkDecorator(ICoffee):
    def __init__(self, coffee: ICoffee):
        self._coffee = coffee

    def get_description(self) -> str:
        # TODO: добавить "с молоком" к описанию внутреннего кофе
        pass

    def get_cost(self) -> float:
        # TODO: стоимость базового кофе + 1$
        pass


# Аналогично реализовать SyrupDecorator, SugarDecorator


# --- Стратегия приготовления (Strategy) ---

class BrewStrategy(ABC):
    @abstractmethod
    def brew(self, coffee: ICoffee) -> str:
        pass


class HotBrew(BrewStrategy):
    def brew(self, coffee: ICoffee) -> str:
        return f"Горячий {coffee.get_description()}"


class IcedBrew(BrewStrategy):
    def brew(self, coffee: ICoffee) -> str:
        return f"Со льдом {coffee.get_description()}"


# Добавить FoamyBrew


# --- Строитель заказов (Builder) ---

class Order:
    def __init__(self):
        self.items: List[ICoffee] = []
        self.delivery_method: Optional[str] = None

    def __str__(self):
        items_str = "\n".join([f"- {item.get_description()} (${item.get_cost():.2f})" for item in self.items])
        return f"Заказ:\n{items_str}\nСпособ доставки: {self.delivery_method or 'не указан'}"


class OrderBuilder:
    def __init__(self):
        self._order = Order()

    def add_coffee(self, coffee: ICoffee):
        self._order.items.append(coffee)

    def set_delivery_method(self, method: str):
        self._order.delivery_method = method

    def get_result(self) -> Order:
        return self._order


# --- Синглтон (CoffeeMachine) ---

class CoffeeMachine:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(CoffeeMachine, cls).__new__(cls)
            cls._instance.orders = []
            cls._instance.clients = []  # для механизма Observer
        return cls._instance

    def prepare_order(self, order: Order):
        print(f"Готовим заказ: {order}")
        self.orders.append(order)
        self._notify_clients(order)

    def subscribe(self, client):
        self.clients.append(client)

    def unsubscribe(self, client):
        self.clients.remove(client)

    def _notify_clients(self, order):
        for client in self.clients:
            client.update(f"Ваш заказ готов: {order}")


# --- Наблюдатель (Observer) ---

class Client:
    def __init__(self, name: str):
        self.name = name

    def update(self, message: str):
        print(f"{self.name}: {message}")


# --- Клиентский код (демонстрация) ---

def run_coffee_system():
    # Шаг 1: Создание напитков через фабрику
    factory = CoffeeFactory()
    espresso = factory.create_espresso()
    cappuccino = factory.create_cappuccino()

    # Шаг 2: Применение декораторов
    espresso_with_milk = MilkDecorator(espresso)
    cappuccino_with_syrup = SyrupDecorator(cappuccino)

    print(f"Напиток: {espresso_with_milk.get_description()}, стоимость: ${espresso_with_milk.get_cost():.2f}")
    print(f"Напиток: {cappuccino_with_syrup.get_description()}, стоимость: ${cappuccino_with_syrup.get_cost():.2f}")

    # Шаг 3: Применение стратегии
    hot_brewer = HotBrew()
    iced_brewer = IcedBrew()

    print(hot_brewer.brew(espresso_with_milk))
    print(iced_brewer.brew(cappuccino_with_syrup))

    # Шаг 4: Сборка заказа через Builder
    builder = OrderBuilder()
    builder.add_coffee(espresso_with_milk)
    builder.add_coffee(cappuccino_with_syrup)
    builder.set_delivery_method("на вынос")
    order = builder.get_result()

    print("\nСоздан заказ:")
    print(order)

    # Шаг 5: Использование Singleton (кофемашина)
    machine = CoffeeMachine()
    client = Client("Иван")
    machine.subscribe(client)
    machine.prepare_order(order)  # должно вывести уведомление для клиента


if __name__ == "__main__":
    run_coffee_system()
```

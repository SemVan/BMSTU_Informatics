```python
from abc import ABC, abstractmethod

# Product
class Transport(ABC):
    @abstractmethod
    def deliver(self):
        pass

class Truck(Transport):
    def deliver(self):
        return "Доставка по суше"

class Ship(Transport):
    def deliver(self):
        return "Доставка по морю"

# Creator
class Logistics(ABC):
    @abstractmethod
    def create_transport(self) -> Transport:
        pass

    def plan_delivery(self):
        transport = self.create_transport()
        return transport.deliver()

class RoadLogistics(Logistics):
    def create_transport(self) -> Transport:
        return Truck()

class SeaLogistics(Logistics):
    def create_transport(self) -> Transport:
        return Ship()

# Клиент
logistics = RoadLogistics()
print(logistics.plan_delivery())  # Доставка по суше
```

```python
class Computer:
    def __init__(self):
        self.cpu = None
        self.ram = None
        self.storage = None

    def __str__(self):
        return f"Компьютер: CPU={self.cpu}, RAM={self.ram}, Storage={self.storage}"

class ComputerBuilder:
    def __init__(self):
        self.computer = Computer()

    def set_cpu(self, cpu):
        self.computer.cpu = cpu
        return self

    def set_ram(self, ram):
        self.computer.ram = ram
        return self

    def set_storage(self, storage):
        self.computer.storage = storage
        return self

    def build(self):
        return self.computer

# Использование без директора
builder = ComputerBuilder()
computer = (builder.set_cpu("Intel i7")
                   .set_ram("16GB")
                   .set_storage("512GB SSD")
                   .build())
print(computer)
```

```python
# Target (ожидаемый интерфейс)
class DataAnalyzer:
    def analyze(self, data):
        pass

# Adaptee (существующий класс с несовместимым интерфейсом)
class JSONAnalyzer:
    def analyze_json(self, json_data):
        return f"Анализ JSON данных: {json_data}"

# Adapter
class CSVToJSONAdapter(DataAnalyzer):
    def __init__(self, json_analyzer: JSONAnalyzer):
        self.json_analyzer = json_analyzer

    def analyze(self, csv_data):
        # Преобразуем CSV в JSON (упрощённо)
        json_data = f'{{"data": "{csv_data}"}}'
        return self.json_analyzer.analyze_json(json_data)

# Клиент
analyzer = CSVToJSONAdapter(JSONAnalyzer())
print(analyzer.analyze("name,age\nAlice,30"))
```

```python
from abc import ABC, abstractmethod

# Component
class Notifier(ABC):
    @abstractmethod
    def send(self, message):
        pass

# ConcreteComponent
class EmailNotifier(Notifier):
    def send(self, message):
        return f"Отправка Email: {message}"

# Base Decorator
class NotifierDecorator(Notifier):
    def __init__(self, notifier: Notifier):
        self._notifier = notifier

    def send(self, message):
        return self._notifier.send(message)

# ConcreteDecorator
class SMSDecorator(NotifierDecorator):
    def send(self, message):
        # Добавляем SMS к отправке
        return super().send(message) + f" + SMS: {message}"

class FacebookDecorator(NotifierDecorator):
    def send(self, message):
        return super().send(message) + f" + Facebook: {message}"

# Использование
notifier = EmailNotifier()
notifier = SMSDecorator(notifier)
notifier = FacebookDecorator(notifier)
print(notifier.send("Привет!"))
# Отправка Email: Привет! + SMS: Привет! + Facebook: Привет!
```

```python
from abc import ABC, abstractmethod

# Strategy
class RouteStrategy(ABC):
    @abstractmethod
    def build_route(self, start, end):
        pass

# ConcreteStrategies
class WalkingRoute(RouteStrategy):
    def build_route(self, start, end):
        return f"Пеший маршрут от {start} до {end} (через парки)"

class CarRoute(RouteStrategy):
    def build_route(self, start, end):
        return f"Автомобильный маршрут от {start} до {end} (по шоссе)"

# Context
class Navigator:
    def __init__(self, strategy: RouteStrategy):
        self._strategy = strategy

    def set_strategy(self, strategy: RouteStrategy):
        self._strategy = strategy

    def build(self, start, end):
        return self._strategy.build_route(start, end)

# Клиент
nav = Navigator(WalkingRoute())
print(nav.build("Парк", "Музей"))
nav.set_strategy(CarRoute())
print(nav.build("Парк", "Музей"))
```

```python
from abc import ABC, abstractmethod

# Observer
class Observer(ABC):
    @abstractmethod
    def update(self, temperature):
        pass

# Subject
class WeatherStation:
    def __init__(self):
        self._observers = []
        self._temperature = 0

    def attach(self, observer: Observer):
        self._observers.append(observer)

    def detach(self, observer: Observer):
        self._observers.remove(observer)

    def notify(self):
        for obs in self._observers:
            obs.update(self._temperature)

    def set_temperature(self, temp):
        self._temperature = temp
        self.notify()

# ConcreteObservers
class CurrentConditionsDisplay(Observer):
    def update(self, temperature):
        print(f"Текущие условия: температура {temperature}°C")

class ForecastDisplay(Observer):
    def update(self, temperature):
        forecast = "тепло" if temperature > 20 else "прохладно"
        print(f"Прогноз: будет {forecast}")

# Клиент
station = WeatherStation()
current = CurrentConditionsDisplay()
forecast = ForecastDisplay()

station.attach(current)
station.attach(forecast)

station.set_temperature(25)
# Текущие условия: температура 25°C
# Прогноз: будет тепло

station.set_temperature(18)
# Текущие условия: температура 18°C
# Прогноз: будет прохладно
```
